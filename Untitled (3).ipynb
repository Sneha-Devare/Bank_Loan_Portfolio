```python
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
import pandas as pd
import plotly.express as px

# create short data
data = {
    "name": ["A", "B", "C", "D"],
    "math": [65, 70, 80, 90],
    "science": [60, 75, 85, 88]
}

df = pd.DataFrame(data)

# plot
fig = px.scatter(
    df,
    x="math",
    y="science",
    text="name",
    title="Math vs Science Marks"
)

fig.show()
```
